```python
import json
import config
from google import genai

client = genai.Client(api_key=config.GEMINI_API_KEY)

print("Gemini client ready.")
print("Model:", config.GEMINI_MODEL)
```

```python
with open("summary.json", "r", encoding="utf-8") as f:
    summary = json.load(f)

print("Summary loaded successfully.")
summary
```

```python
prompt = f"""
You are a sharp, professional business analyst for a beauty and makeup ecommerce business.

Write a concise weekly business report using only the data provided below.
Do not invent numbers.
Do not hallucinate.
Do not mention a person's name.
Do not use starred markdown formatting.
Do not use hashtags.
Do not write a long essay.
Do not copy the wording from any sample.
Instead, follow the same reporting format and make it specific to this business.

DATA:
{json.dumps(summary, indent=2)}

Write the output in this exact structure:

AI Summary
- Write 1 short opening statement about the week overall.
- Keep it direct, polished, and business-focused.

Executive Summary
- Write 3 to 4 bullet points.
- Explain the main weekly performance changes.
- Mention revenue, orders, AOV, and customer mix.
- Keep each bullet short and readable.

Key Metrics
- Write 4 to 6 bullet points.
- Include current week revenue, previous week revenue, revenue change, orders, previous orders, orders change, AOV, previous AOV, AOV change, customer split, and leading channel.
- Use simple business language.

Top-Performing Products
- Write 3 bullet points.
- Mention the top products by revenue from the available data.
- Keep the wording factual and clear.

Notable Anomalies
- Write 1 to 2 bullet points.
- If anomaly_days_count is 0, say that no unusual spikes or drops were flagged in the weekly data.
- If anomalies exist, summarize them briefly and clearly.

Actionable Recommendations
- Write 3 bullet points.
- Give practical next-step recommendations for the business owner.
- Focus on revenue recovery, channel optimization, product performance, and operational actions.

Charts
- Write 1 to 2 bullet points.
- Mention these visuals by name and explain that they help interpret weekly performance:
  Revenue Week over Week, Daily Revenue Trend, Top Products by Revenue, Revenue Share by Channel, and Orders Week over Week.

Style rules:
- Use bullet points for readability.
- Keep the tone polished, bold, and presentable.
- Make the report easy to scan.
- Use short, business-focused language.
- Do not use starred bullet formatting.
- Do not add any sections outside the ones listed above.
"""
```

```python
response = client.models.generate_content(
    model=config.GEMINI_MODEL,
    contents=prompt
)

llm_summary = response.text.strip()
print(llm_summary)
```

```python
print(type(llm_summary))
print(llm_summary[:1500])
```

```python
clean_report = f"""Weekly Makeup Business Report

{llm_summary}
"""

with open("weekly_report.md", "w", encoding="utf-8") as f:
    f.write(clean_report)

with open("weekly_report.txt", "w", encoding="utf-8") as f:
    f.write(clean_report)

print("weekly_report.md and weekly_report.txt saved successfully.")
```
