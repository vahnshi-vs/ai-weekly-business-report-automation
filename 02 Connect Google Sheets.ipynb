```python
import gspread
from google.oauth2.service_account import Credentials
import pandas as pd

# Scopes tell Google which APIs we are allowed to use
SCOPES = [
    "https://www.googleapis.com/auth/spreadsheets.readonly",
    "https://www.googleapis.com/auth/drive.readonly",
]

# Load credentials from your JSON key file
creds = Credentials.from_service_account_file(
    "service_account.json",
    scopes=SCOPES
)
client = gspread.authorize(creds)

# TODO: Paste your Sheet ID here between the quotes
SHEET_ID = "1bRQhejMBoDTXLmT1qGlK1J7Fyalp9htb1valQ6iObDs"

# TODO: Paste your worksheet/tab name here exactly
WORKSHEET_NAME = "raw_sales_data"

# Open the sheet and read all rows
sheet = client.open_by_key(SHEET_ID)
worksheet = sheet.worksheet(WORKSHEET_NAME)
rows = worksheet.get_all_records()

# Convert to DataFrame
df_raw = pd.DataFrame(rows)

# Write and print success message
message = "Connection successful"
print(message)
print("Spreadsheet title:", sheet.title)
print("Worksheet title:", worksheet.title)
print("Shape:", df_raw.shape)

display(df_raw.head())
```
